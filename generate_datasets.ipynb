```python
import pandas as pd
import numpy as np
```

```python
# read tsv
df5comb = pd.read_csv('doi_10_5061_dryad_sm847__v20181016/db_drugs_5s.tsv', sep='\t')
df4comb = pd.read_csv('doi_10_5061_dryad_sm847__v20181016/db_drugs_4s.tsv', sep='\t')
df3comb = pd.read_csv('doi_10_5061_dryad_sm847__v20181016/db_drugs_3s.tsv', sep='\t')
```

```python
# get drug_name_A, drug_name_B, and drug_name_C to create a new column
# array 
df5comb['drug_names'] = df5comb[['drug_name_A', 'drug_name_B', 'drug_name_C', 'drug_name_D', 'drug_name_E']].values.tolist()
df4comb['drug_names'] = df4comb[['drug_name_A', 'drug_name_B', 'drug_name_C', 'drug_name_D']].values.tolist()
df3comb['drug_names'] = df3comb[['drug_name_A', 'drug_name_B', 'drug_name_C']].values.tolist()

# concat all the datasets to one (before it, get only drug_names, exact_exposure_count, atleast_exposure_count columns
df5comb = df5comb[['drug_names', 'exact_exposure_count', 'atleast_exposure_count']]
df4comb = df4comb[['drug_names', 'exact_exposure_count', 'atleast_exposure_count']]
df3comb = df3comb[['drug_names', 'exact_exposure_count', 'atleast_exposure_count']]

dfcomb = pd.concat([df5comb, df4comb, df3comb])
```

```python
# get most common 40 drugs
most40common = dfcomb['drug_names'].explode().value_counts().head(100).index.tolist()
most40common
```

```python
# filter each drug_names array so that it contains only the most common 40 drugs
dfcomb_filter40 = dfcomb.copy()
dfcomb_filter40['drug_names'] = dfcomb_filter40['drug_names'].apply(lambda x: [i for i in x if i in most40common])
dfcomb_filter40 = dfcomb_filter40[dfcomb_filter40['drug_names'].apply(lambda x: len(x) >= 3)]
```

```python
import pandas as pd
```

```python
df = pd.read_sas('RXQ_RX_G.XPT', format='xport', encoding='utf-8')
df
```

```python
top40 = df['RXDDRUG'].value_counts().head(40).index.tolist()[1:]

# Keep only top 40 drugs
filtered = df[df['RXDDRUG'].isin(top40)]

# Group by column SEQN and aggregate RXDDRUG
filtered = filtered.groupby('SEQN')['RXDDRUG'].apply(list).reset_index()

# remove less than 3 drugs in combination
filtered = filtered[filtered['RXDDRUG'].apply(len) >= 3]
```

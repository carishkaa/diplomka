```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('refined_drug_data.csv', sep=';')
df = df.fillna('')

dosages = pd.read_csv('drugs_dosing.csv', sep=';')
dosages = dosages[['Drug Names', 'MIN units [pcs]', 'MAX units [pcs]']]
```

'Dosage_1', 'Dosage_2', 'Dosage_3', 'Dosage_4', 'Dosage_5', 'Dosage_6', 'Dosage_7', 'Dosage_8' columns will contain randomly generated dosages (number of units) for each drug.

Processing speed of dispensing machine is 10 tablets per second, so later we can calculate the time required to dispense each drug by dividing the dosage by 10.

```python
for i in range(1, 9):
    df['Dosage_' + str(i)] = 0

for index, row in df.iterrows():
    for i in range(1, 9):
        if row['Drug_' + str(i)] != '':
            drug_name = row['Drug_' + str(i)]
            min_dosage = dosages[dosages['Drug Names'] == drug_name]['MIN units [pcs]'].values[0]
            max_dosage = dosages[dosages['Drug Names'] == drug_name]['MAX units [pcs]'].values[0]
            df.at[index, 'Dosage_' + str(i)] = round(np.random.randint(min_dosage, max_dosage))
df = df.replace(0, '')
df
```

```python
df.to_csv('generated_data.csv', sep=';', index=False)
```
